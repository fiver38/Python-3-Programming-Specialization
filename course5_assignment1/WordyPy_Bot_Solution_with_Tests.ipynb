
# Assignment 1: Shall we play a game?
By using Object-Oriented Programming (OOP), you will create a program called **WordyPy**. In WordyPy, the bot you create must try to guess a five-letter secret word based on feedback given from the game engine. 

The main tasks:
1. Implement the **Letter** class.
2. Implement the **Bot** class with `make_guess` and `record_guess_results` methods.
3. Use the provided GameEngine to test your bot.

---


```python

class Letter:
    def __init__(self, letter: str, index: int):
        self.letter = letter
        self.index = index
        self.in_word = False
        self.correct_place = False

    def set_in_word(self, value: bool):
        self.in_word = value

    def set_correct_place(self, value: bool):
        self.correct_place = value

    def is_in_word(self) -> bool:
        return self.in_word

    def is_in_correct_place(self) -> bool:
        return self.correct_place

    def __repr__(self):
        return f"Letter({self.letter!r}, {self.index}, in_word={self.in_word}, correct_place={self.correct_place})"
```

```python

import random

class Bot:
    def __init__(self, word_list_file: str):
        with open(word_list_file, 'r') as f:
            self.word_list = [w.strip().upper() for w in f.readlines() if len(w.strip()) == 5]
        self.guess_list = []
        self.letter_list = []

    def make_guess(self) -> str:
        if not self.letter_list:
            guess = random.choice([w for w in self.word_list if w not in self.guess_list])
        else:
            candidates = [w for w in self.word_list if w not in self.guess_list]
            for l in self.letter_list:
                if l.is_in_correct_place():
                    candidates = [w for w in candidates if w[l.index] == l.letter]
                elif l.is_in_word():
                    candidates = [w for w in candidates if l.letter in w]
            guess = random.choice(candidates) if candidates else random.choice(self.word_list)
        self.guess_list.append(guess)
        return guess

    def record_guess_results(self, guess: str, feedback: str):
        for i, symbol in enumerate(feedback):
            letter_obj = Letter(guess[i], i)
            if symbol == "*":
                letter_obj.set_in_word(True)
            if symbol == "?":
                pass
            if symbol != "?":
                if symbol == guess[i]:
                    letter_obj.set_in_correct_place(True)
                    letter_obj.set_in_word(True)
            self.letter_list.append(letter_obj)
```

```python

# Example GameEngine for testing

class GameEngine:
    def __init__(self, target_word: str, bot: Bot):
        self.target = target_word.upper()
        self.bot = bot

    def evaluate_guess(self, guess: str) -> str:
        feedback = ""
        for i, char in enumerate(guess):
            if char == self.target[i]:
                feedback += char
            elif char in self.target:
                feedback += "*"
            else:
                feedback += "?"
        return feedback

    def play(self):
        for attempt in range(6):
            guess = self.bot.make_guess()
            feedback = self.evaluate_guess(guess)
            print(f"Guess {attempt+1}: {guess} -> {feedback}")
            if guess == self.target:
                print("Correct! You win!")
                return
            self.bot.record_guess_results(guess, feedback)
        print(f"Failed! Target word was {self.target}.")

# Example run
word_list_file = "words.txt"  # Replace with your word list path
bot = Bot(word_list_file)
engine = GameEngine("APPLE", bot)
engine.play()
```
