```
# ✅ Función lambda (función anónima inline)
sumar = lambda x, y: x + y  # Suma dos números
print(sumar(3, 5))  # Resultado: 8

# ✅ Condición inline (operador ternario)
edad = 20
mensaje = "Mayor" if edad >= 18 else "Menor"  # Evalúa condición en una sola línea
print(mensaje)  # Resultado: Mayor

# ✅ Asignación inline (operador walrus :=) – desde Python 3.8+
# Asigna y evalúa al mismo tiempo dentro de una expresión
if (n := len("hola mundo")) > 5:
    print(n)  # Resultado: 10

# ✅ List comprehension (crea listas en una línea)
cuadrados = [x**2 for x in range(5)]  # Genera lista de cuadrados del 0 al 4
print(cuadrados)  # Resultado: [0, 1, 4, 9, 16]

# ✅ Set comprehension (crea conjuntos en una línea)
pares = {x for x in range(10) if x % 2 == 0}
print(pares)  # Resultado: {0, 2, 4, 6, 8}

# ✅ Dict comprehension (crea diccionarios en una línea)
cuadrados_dict = {x: x**2 for x in range(4)}  # Clave: valor al cuadrado
print(cuadrados_dict)  # Resultado: {0: 0, 1: 1, 2: 4, 3: 9}

# ✅ map con lambda (transformar lista en una línea)
nums = [1, 2, 3]
dobles = list(map(lambda x: x * 2, nums))  # Duplicar cada elemento
print(dobles)  # Resultado: [2, 4, 6]

# ✅ filter con lambda (filtrar lista en una línea)
nums = [1, 2, 3, 4]
solo_pares = list(filter(lambda x: x % 2 == 0, nums))  # Filtrar pares
print(solo_pares)  # Resultado: [2, 4]

# ✅ Unpacking (asignación múltiple en una línea)
a, b = [10, 20]
print(a + b)  # Resultado: 30

# ✅ Condiciones múltiples inline (encadenamiento ternario)
nota = 85
estado = "Excelente" if nota >= 90 else "Bueno" if nota >= 70 else "Insuficiente"
print(estado)  # Resultado: Bueno
```
